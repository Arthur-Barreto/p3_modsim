```python
# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from math import *
from scipy.integrate import odeint
import pandas
```

```python
# Implemente seu código do item 0a abaixo
r = 5e-2 # em m
p = 161.5# em kg/m³

g = 9.81 # em m/s²
pa = 1.29 # em kg/m³
Cd = 48e-2 # adimencional
# A seguir temos a área da secção transversal e o volume da esfera
A = pi*r**2 # em m²
V = (4/3)*pi*r**3 # em m³

m = p*V
```

```python
# Implemente seu código do item 0b abaixo
# primeiro as constantes que foi fornecido
v0 = 3.49035865684291
teta = 56.65929265099678
# convertendo o angulo em radianos
teta = radians(teta)
# agora as componentes das velocidades
v0x = v0*cos(teta)
v0y = v0*sin(teta)

#agora a nossa lista de tempo

lisTempo = np.arange(0,5,1e-3)
```

```python
def modelo(lista,t):
    # definindo os parametros com o vetor x
    x = lista[0] 
    y = lista[1] 
    vx = lista[2] 
    vy = lista[3] 
    # definindo as derivadas
    
    v = sqrt(vx**2+vy**2)
    
    dxdt = vx
    dydt = vy
    dvxdt = (1/m)*(-0.5*pa*A*Cd*vx*v)
    dvydt = (1/m)*(-m*g -0.5*pa*A*Cd*vy*v)
    # tratando o caso que ela atinge o chão
    if y < 0:
        dxdt = 0
        dydt = 0
        dvxdt = 0
        dvydt = 0
    # concatenando as derivadas em um vetor
    info = [dxdt,dydt,dvxdt,dvydt]
    # retorando o vetor
    return info
```

```python
# rodando odeint
# definindo as condições iniciais
ci = [0,0,v0x,v0y]

# criando a lista de tempo
lisTempo = np.arange(0,10,1e-3)
print(len(lisTempo))

#executando odeint
info = odeint(modelo,ci,lisTempo)
x = info[:,0]
y = info[:,1]

plt.plot(x, y, label = "Teste1")
plt.axis('equal')
plt.title("Primeira iteração")
plt.ylabel("Altura (m)")
plt.xlabel("Espaço (m)")
plt.legend()
plt.grid(True)
plt.show()
```

# segunda iteração


```python
# considerando a força de magnus
# fmg = 0.5*pa*A*Cm*v*w, onde w é velocidade angular e Cm é o coeficiente de magnus
# também vamos considerar o empuxo do ar

def modelo1(lista,t,Cm):
    # definindo os parametros com o vetor x
    x = lista[0] 
    y = lista[1] 
    vx = lista[2] 
    vy = lista[3] 
    
    # valores auxiliares
    v = sqrt(vx**2+vy**2)
    w = v/r
    # definindo as derivadas
    
    dxdt = vx
    dydt = vy
    dvxdt = (1/m)*(-0.5*pa*A*Cd*vx*v + 0.5*pa*A*Cm*vy*v)
    dvydt = (1/m)*(-m*g -0.5*pa*A*Cd*vy*v -0.5*pa*A*Cm*v*vx + pa*V*g)
    # tratando o caso que ela atinge o chão
    if y < 0:
        dxdt = 0
        dydt = 0
        dvxdt = 0
        dvydt = 0
    # concatenando as derivadas em um vetor
    info = [dxdt,dydt,dvxdt,dvydt]
    # retorando o vetor
    return info
```

# validação do modelo

```python
lisCm = np.arange(10,25,1)
for Cm in lisCm:
    info1 = odeint(modelo1,ci,lisTempo,args=(Cm,))
    x1 = info1[:,0]
    y1 = info1[:,1]
    plt.plot(x1, y1)
    
plt.plot(xe,ye, "mo", label = "Experimento")
plt.axis('equal')
plt.title("Segunda iteração")
plt.ylabel("Altura (m)")
plt.xlabel("Espaço (m)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# modelo validado
# vamos executar a odeint para varias densidades

def modelo2(lista,t,d):
    # definindo os parametros com o vetor x
    x = lista[0] 
    y = lista[1] 
    vx = lista[2] 
    vy = lista[3] 
    
    M = d*V
    
    # valores auxiliares
    v = sqrt(vx**2+vy**2)
    w = v/r
    # definindo as derivadas
    
    dxdt = vx
    dydt = vy
    dvxdt = (1/M)*(-0.5*pa*A*Cd*vx*v + 0.5*pa*A*Cm*vy*v)
    dvydt = (1/M)*(-M*g -0.5*pa*A*Cd*vy*v -0.5*pa*A*Cm*v*vx + pa*V*g)
    # tratando o caso que ela atinge o chão
    if y < 0:
        dxdt = 0
        dydt = 0
        dvxdt = 0
        dvydt = 0
    # concatenando as derivadas em um vetor
    info = [dxdt,dydt,dvxdt,dvydt]
    # retorando o vetor
    return info
```

```python
# criando uma lista de densidades
Cm = 22

densidades = np.arange(161.5,10000,10)
x_max = []
y_max = []

for densidade in densidades:
    info2 = odeint(modelo2,ci,lisTempo, args=(densidade,))
    x2 = info1[:,0]
    y2 = info1[:,1]
    x_max.append(max(x2))
    y_max.append(max(y2))
    plt.plot(x2,y2)  

print(len(x_max))
print(len(y_max))

plt.title("Influencia da densidade")
plt.ylabel("Altura (m)")
plt.xlabel("Espaço (m)")
plt.grid(True)
plt.show()
```
